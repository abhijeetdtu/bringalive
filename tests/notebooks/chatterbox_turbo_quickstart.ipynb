# Chatterbox Turbo Quick Test (llm3)
Uses chatterbox source install and ChatterboxTurboTTS.


```python
import os
import torch
import torchaudio
from chatterbox.tts_turbo import ChatterboxTurboTTS

print('torch:', torch.__version__)

device = 'mps' if torch.backends.mps.is_available() else 'cpu'
```

```python
device
```

```python
model = ChatterboxTurboTTS.from_pretrained(device=device)
```

```python
model.prepare_conditionals(
    "/Users/abhijeetpokhriyal/Downloads/voice_clone.wav",
    exaggeration=0.3
)
```

```python
text = open("./output.txt" , "r").read()
text[:20]
```

```python
# text = 'Hello from Chatterbox Turbo on Apple Silicon.'
wav = model.generate(
    text[:350],
    # temperature=0.4,
    # top_p=0.9,
    # repetition_penalty=1.1
)
```

```python
# Save output
out_path = 'chatterbox_turbo_test.wav'
wav_out = wav.squeeze()
if wav_out.dim() == 1:
    wav_out = wav_out.unsqueeze(0)
torchaudio.save(out_path, wav_out.cpu(), 24000)
print('Saved:', os.path.abspath(out_path))
```
