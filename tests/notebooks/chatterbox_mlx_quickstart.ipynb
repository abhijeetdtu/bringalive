# Chatterbox (MLX) Quick Test
This notebook loads Chatterbox MLX and generates a short sample using Apple Silicon acceleration.


```python
import os
import torch
import torchaudio
from chatterbox.tts import ChatterboxTTS


print('torch:', torch.__version__)

# MLX devices typically use Apple Silicon; 'hybrid-mlx' is recommended
# device = 'hybrid-mlx'
```

```python
torch.backends.mps.is_available()
```

```python
device= "mps"
model = ChatterboxTTS.from_pretrained(device=device)
```

```python
model.prepare_conditionals(
    "/Users/abhijeetpokhriyal/Downloads/voice_clone.wav",
    exaggeration=0.7
)
```

```python
text = open("./output.txt" , "r").read()
text[:20]
```

```python
# text = 'Hello from Chatterbox MLX on Apple Silicon. This is a quick test.'
wav = model.generate(
    text,
    # audio_prompt_path="/Users/abhijeetpokhriyal/Downloads/voice_clone_2.wav",
    cfg_weight=0.3,
    temperature=0.4,
    top_p=0.95,
    repetition_penalty=1.1,
)
```

```python
# Save output
out_path = 'chatterbox_test.wav'
wav_out = wav.squeeze()
if wav_out.dim() == 1:
    wav_out = wav_out.unsqueeze(0)
# torchaudio expects [channels, samples]
torchaudio.save(out_path, wav_out.cpu(), 24000)
print('Saved:', os.path.abspath(out_path))
```
